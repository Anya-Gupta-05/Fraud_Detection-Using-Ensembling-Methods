```python
#cleaning the data 
#removing unwanted columns 
df = df.drop(
    columns=['Transaction ID', 'Customer ID', 'Customer Location','Transaction Date','IP Address', 'Shipping Address', 'Billing Address'], 
    errors='ignore')
```

```python
cat_columns=df.select_dtypes(include=['category','object']).columns
print("Categorical Columns are:",cat_columns.tolist())
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.columns=df.columns.str.strip()
print(repr(df.columns))
```

```python
df.columns = df.columns.str.encode('ascii', 'ignore').str.decode('ascii')
for col in cat_columns:
    if col in df.columns:
        df[col] = le.fit_transform(df[col])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,accuracy_score
```

```python
X=df.drop('Is Fraudulent',axis=1)
y=df['Is Fraudulent']
X.shape,y.shape
```

```python
x_sub=X[:30000]
y_sub=y[:30000]
```

```python
X_train,X_test,y_train,y_test=train_test_split(x_sub,y_sub,test_size=0.2,random_state=99)
rf=RandomForestClassifier(random_state=99)
para_grid={
     'n_estimators':[100,200,300],
     'max_depth':[None,10,20],
     'min_samples_split':[2,5,10]
}
```

```python
rf=RandomForestClassifier(random_state=99)
grid_search=GridSearchCV(estimator=rf,param_grid=para_grid,cv=3,scoring='accuracy')
grid_search.fit(x_sub,y_sub)
```

```python
best_params=grid_search.best_params_
print(f"the best parametrs are:{best_params}")
```

```python
rffinal=RandomForestClassifier(random_state=99,**best_params)
rffinal.fit(x_sub,y_sub)
```

```python
y_pred=rffinal.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print(f"the accuracy In Random Forest is:{accuracy*100}")
```

```python
from xgboost import XGBClassifier
param_gridxgb = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 10],
    'learning_rate': [0.01, 0.1, 0.2],
    'min_child_weight': [1, 5, 10],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
}
from sklearn.model_selection import GridSearchCV
xgb_model = XGBClassifier(random_state=99)
xgb_grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_gridxgb, cv=3, scoring='accuracy')
xgb_grid_search.fit(x_sub, y_sub)
xgb_best_params = xgb_grid_search.best_params_
print("Best Parameters:", xgb_best_params)
```

```python
xgb_final = XGBClassifier(random_state=99, **xgb_best_params)
xgb_final.fit(x_sub, y_sub)
y_xgb_predict=xgb_final.predict(X_test)
xgb_accuracy=accuracy_score(y_test,y_xgb_predict)
print(f"the accuracy In XGBoost Classifier is:{xgb_accuracy*100}")
```

```python
#Voting Classification for both the models 
#Performing Soft Voting 
```

```python
from sklearn.ensemble import VotingClassifier
vot_cl=VotingClassifier(
    estimators=[('rf',rffinal),('xgb',xgb_final)],
    voting='soft')
```

```python
vot_cl.fit(X_train,y_train)
vot_predict=vot_cl.predict(X_test)
acc=accuracy_score(y_test,vot_predict)
print("Model Accuracy after Soft Voting of Random Forest and XGBoost Classifier:",acc*100)
```

```python
vot_cl2=VotingClassifier(estimators=[('rf',rffinal),('xgb',xgb_final)],voting='hard')
vot_cl2.fit(X_train,y_train)
vot_predict_hard=vot_cl2.predict(X_test)
acc_hard=accuracy_score(y_test,vot_predict_hard)
print("Model Accuracy after Hard Voting of Random Forest and XGBoost Classifier:",acc_hard*100)
```
